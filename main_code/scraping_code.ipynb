#### **Scraping from Google Play Scraper**

Scraping was performed using the 'google_play_scraper' library. The data was collected based on the application's package name, in this case, reviews of **Mobile Legends Bang Bang**. A total of 100k reviews were retrieved from the United States and sorted by relevance.

For 'google_play_scraper' documentation: https://pypi.org/project/google-play-scraper/

```python
from google_play_scraper import reviews, Sort
import pandas as pd

app_id = "com.mobile.legends" 

result, _ = reviews(
    app_id,
    lang="en",
    country="us", 
    sort=Sort.MOST_RELEVANT, 
    count=100000, 
    filter_score_with=None
)

df = pd.DataFrame(result)

df.to_csv(r"c:\Users\USER\Documents\Belajar Data Science dicoding\Belajar Pengembangan Machine Learning\submission\output\scraping_review_MLBB.csv", index=False)

df.head()
```
